```python
df_selected = df_year[df_year["indicator_name"].isin(selected_indicators)].copy()

print(df_selected.shape)
df_selected[["name", "region", "year", "indicator_name", "indicator_value"]].head(15)
```

```python
df_selected["indicator_value"] = (
    df_selected["indicator_value"]
    .astype(str)
    .str.replace("\xa0", "", regex=False)
    .str.replace(" ", "", regex=False)
    .str.replace(",", ".", regex=False)
)

df_selected["indicator_value"] = pd.to_numeric(df_selected["indicator_value"], errors="coerce")

df_selected[["indicator_name", "indicator_value"]].head(10)
```

```python
duplicates = (
    df_selected
    .groupby(["id", "name", "region", "year", "indicator_name"])
    .size()
    .reset_index(name="count")
)

duplicates[duplicates["count"] > 1].head(20)
```

```python
df_wide = df_selected.pivot_table(
    index=["id", "name", "region", "year"],
    columns="indicator_name",
    values="indicator_value",
    aggfunc="first"
).reset_index()

print(df_wide.shape)
df_wide.head()
```

```python
df_wide = df_wide.rename(columns={
    "Средний  ЕГЭ студентов принятых на обучение по программам бакалавриата и специалитета по всем формам обучения": "ege_avg",
    "Приведенный контингент студентов": "students_count",
    "Доля иностранных студентов в общей численности студентов, обучающихся по программам бакалавриата, специалитета, магистратуры": "foreign_share",
    "Средняя заработная плата ППС (без внешних совместителей и работающих по договорам ГПХ)": "pps_salary",
    "Общее количество публикаций организации в расчете на НПР": "publications_per_staff",
    "Количество цитирований публикаций, изданных за последние лет, индексируемых в Российском индексе научного цитирования (далее – РИНЦ) в расчете на НПР[н]": "citations_per_staff",
    "Доходы образовательной организации из всех источников в расчете на численность студентов (приведенный контингент)": "income_per_student"
})

df_wide.head()
```

```python
cluster_df = df_wide.dropna(subset=[
    "ege_avg",
    "students_count",
    "foreign_share",
    "pps_salary",
    "publications_per_staff",
    "citations_per_staff",
    "income_per_student"
]).copy()

print(cluster_df.shape)
cluster_df.head()
```

```python
feature_cols = [
    "ege_avg",
    "students_count",
    "foreign_share",
    "pps_salary",
    "publications_per_staff",
    "citations_per_staff",
    "income_per_student"
]

cluster_df[feature_cols].describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(cluster_df[feature_cols])

X_scaled[:5]
```

```python
from sklearn.cluster import KMeans

inertias = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(list(k_values), inertias, marker="o")
plt.xlabel("Количество кластеров k")
plt.ylabel("Inertia")
plt.title("Метод локтя")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

for k, score in zip(k_values, silhouette_scores):
    print(f"k = {k}: silhouette = {score:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(list(k_values), silhouette_scores, marker="o")
plt.xlabel("Количество кластеров k")
plt.ylabel("Silhouette score")
plt.title("Выбор числа кластеров по silhouette score")
plt.grid(True)
plt.show()
```

```python
best_k = 4

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_df["cluster"] = kmeans.fit_predict(X_scaled)

cluster_df[["name", "region", "cluster"] + feature_cols].head()
```

```python
cluster_df["cluster"].value_counts().sort_index()
```

```python
cluster_profile = cluster_df.groupby("cluster")[feature_cols].mean().round(2)
cluster_profile
```

```python
cluster_df.to_csv("universities_clustered_2025.csv", sep=";", index=False)
cluster_profile.to_csv("cluster_profile_2025.csv", sep=";")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

cluster_df["pca1"] = X_pca[:, 0]
cluster_df["pca2"] = X_pca[:, 1]

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    cluster_df["pca1"],
    cluster_df["pca2"],
    c=cluster_df["cluster"],
    alpha=0.7
)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Кластеры вузов в пространстве PCA")
plt.grid(True)
plt.show()
```

```python
cluster_df_filtered = cluster_df[
    (cluster_df["students_count"] >= 20) &
    (cluster_df["ege_avg"] > 0)
].copy()

print(cluster_df.shape)
print(cluster_df_filtered.shape)
```

```python
import numpy as np

cluster_df_log = cluster_df_filtered.copy()

log_cols = [
    "students_count",
    "publications_per_staff",
    "citations_per_staff",
    "income_per_student"
]

for col in log_cols:
    cluster_df_log[col] = np.log1p(cluster_df_log[col])

cluster_df_log[feature_cols].head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled_log = scaler.fit_transform(cluster_df_log[feature_cols])

X_scaled_log[:5]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

k_values = range(2, 11)
silhouette_scores_log = []
inertias_log = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled_log)
    silhouette_scores_log.append(silhouette_score(X_scaled_log, labels))
    inertias_log.append(kmeans.inertia_)

for k, score in zip(k_values, silhouette_scores_log):
    print(f"k = {k}: silhouette = {score:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(list(k_values), inertias_log, marker="o")
plt.xlabel("Количество кластеров k")
plt.ylabel("Inertia")
plt.title("Метод локтя после фильтрации и логарифмирования")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(list(k_values), silhouette_scores_log, marker="o")
plt.xlabel("Количество кластеров k")
plt.ylabel("Silhouette score")
plt.title("Silhouette score после фильтрации и логарифмирования")
plt.grid(True)
plt.show()
```

```python
best_k = 3

kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_df_log["cluster"] = kmeans_final.fit_predict(X_scaled_log)

print(cluster_df_log["cluster"].value_counts().sort_index())

cluster_profile_final = cluster_df_log.groupby("cluster")[feature_cols].mean().round(2)
cluster_profile_final
```
